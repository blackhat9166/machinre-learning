```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
flw = pd.read_csv('Iris.csv')
flw.head()
```

```python
flw.rename(columns={'SepalLengthCm':'sl','SepalWidthCm':'sw','PetalLengthCm':'pl','PetalWidthCm':'pw'},inplace=True)
flw
```

```python
setosa = flw[flw['Species'] == 'Iris-setosa']
versicolor= flw[flw['Species'] == 'Iris-versicolor']
virginica = flw[flw['Species'] == 'Iris-virginica']
```

1. PDF  (probability distribution function)
2. CDF  (Cumulative Distribution Function)

```python
#pdf is probability of a particular value in the dataset
#cdf is the probability of all value which is less than the given value
freq ,sp = np.histogram(setosa['pl'],bins=10)
print(freq)
print(sp)
```

```python
pdf = freq/np.sum(freq) # this is pdf of each
```

```python
cdf = np.cumsum(pdf)  #this our cdf
cdf
```

```python
plt.plot(sp[1:],pdf)
plt.plot(sp[1:],cdf)
```

```python
#now lets have the cdf and  pdf of  all three pl
freq ,sp = np.histogram(setosa['pl'],bins=10)
pdf = freq/np.sum(freq)
cdf = np.cumsum(pdf)  
plt.plot(sp[1:],pdf)
plt.plot(sp[1:],cdf)

freq ,sp = np.histogram(versicolor['pl'],bins=10)
pdf = freq/np.sum(freq)
cdf = np.cumsum(pdf)  
plt.plot(sp[1:],pdf)
plt.plot(sp[1:],cdf)

freq ,sp = np.histogram(virginica['pl'],bins=10)
pdf = freq/np.sum(freq)
cdf = np.cumsum(pdf)  
plt.plot(sp[1:],pdf)
plt.plot(sp[1:],cdf)

```
