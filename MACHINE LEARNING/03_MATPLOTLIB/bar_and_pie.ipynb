to see three bars of each side by side than we require numpy array

```python
ages = np.arange(len(age_x))
width = 0.25
plt.bar(ages-width,sal_y,color='#ada4a2',width=width,label='javac')
plt.bar(ages,sal2_y ,width=width,label='pythoonn')
plt.bar(ages+width,sal_y,width=width, label='Javac')
plt.legend()
```

```python
# to have the ages back on the axes we use xticks
ages = np.arange(len(age_x))
width = 0.25
plt.bar(ages-width,sal_y,color='#ada4a2',width=width,label='javac')
plt.bar(ages,sal2_y ,width=width,label='pythoonn')
plt.bar(ages+width,sal_y,width=width, label='Javac')
plt.xticks(ticks= ages,labels= age_x)
plt.legend()
```

```python
data = pd.read_csv('lg.csv')
```

TO HANDLE THIS WE USE COUNTER OF PYTHON LIBRARY FROM COLLECTIONS


```python
#it is used make a count of every unique value in list over iteration of multiple list
from collections import Counter
d =  Counter(['python', 'java', 'python'])
d
d.update(['java','java','java','java','java','java'])
d
```

```python
x = data["LanguagesWorkedWith"].str.split(';').tolist()
final = Counter(x[0])

for i in range(1,87569):
   final.update(x[i])

final.most_common(20)  # most common is used for take major 20 
lang = []
pop = []

for i in final.most_common(20):
   lang.append(i[0])
   pop.append(i[1])
print(pop)
print(lang)


```

```python
plt.barh(lang,pop)
plt.xlabel('popularity')
plt.ylabel('language')
plt.title('pop of languages')
```
