```python
title = []
fulltitle = []
type = []
year = [] 
imDB = []
```

```python
for i in range(100000,300001):
    res =requests.get('https://imdb-api.com/API/Ratings/k_6oiz56np/tt0{}'.format(i)).json()
    title.append(res['title'])
    fulltitle.append(res['fullTitle'])
    type.append(res['type'])
    year.append(res['year'])
    imDB.append(res['imDb'])
    
dic = {'title':title, 'fulltitle':fulltitle, 'type':type, 'year':year, 'imDB':imDB}
```

```python

final = pd.DataFrame.from_dict(dic, orient='index')
final = final.transpose()
final.head(60)
```
