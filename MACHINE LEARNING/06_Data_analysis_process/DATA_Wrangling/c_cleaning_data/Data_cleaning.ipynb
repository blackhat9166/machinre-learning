lets start:
a. completeness issues
  -  missing values for contact details 12[cmp]
  - only 280 rows are there instead of 350
  - missing value in hba1c_change


```python
import pandas as pd

patients = pd.read_csv('insulin_patient.csv')
treatment = pd.read_csv('treatment.csv')
adverse = pd.read_csv('adverse_eff.csv')
treatment_cut = pd.read_csv('treatment_cut.csv')
```

```python
# always create copy
patients_copy = patients.copy()
treatment_copy = treatment.copy()
adverse_copy = adverse.copy()
treatment_cut_copy = treatment_cut.copy()
```

```python
# we concate the two csv to form 350 complete treatment sheet
treatment_f = pd.concat([treatment_copy,treatment_cut_copy],ignore_index= True)
```

```python
# we will hba1c solve
# hba1c_start - hb1ac_end = hb1ac_change
treatment_f['hba1c_change'] = treatment_f['hba1c_start']-treatment_f['hba1c_end']
```

```python
treatment_f['hba1c_change'].isnull().sum()
```

b. Tidiness issue

1. phone email both in same column
2. novodra and aurolin col
3. merge adverse table to treatment table

```python
#novodra and aurolin using melt functions

treatment_f = treatment_f.melt(id_vars=['given_name','surname','hba1c_start','hba1c_end','hba1c_change'],
value_vars=['auralin','novodra'],
var_name='treatmen',value_name='dose')

# here  id_var = list contains column jiske about baki column ko ghumana he
#    value_var = jisko poora bichana he har ek row me.... agar koi column pass ni kiya to jo id_var me ni he wo sare ho jayenge
#  var_name = jo column poore me bichh raha he uske column ka name
#  var_values = bichne wale column ki jitni value he
treatment_f
```

```python
treatment_f = treatment_f[treatment_f['dose'] != '-']
treatment_f
```

```python
treatment_f['start'],treatment_f['end'] = treatment_f['dose'].str.split('-').str
treatment_f = treatment_f.drop(columns='dose')
```

```python
treatment_f['start'] =treatment_f['start'].str.extract('(\d+)').astype(int)
treatment_f['end'] =treatment_f['end'].str.extract('(\d+)').astype(int)

```
