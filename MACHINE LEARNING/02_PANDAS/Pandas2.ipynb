Q - most destructive over (16-20) scorer batsman in history

```python
# find the max strike rate = (no. of rates/no. of balls )*100
# min batsman balls in over 16-20

death_over = delivery[delivery['over'] > 15]
x = death_over.groupby('batsman')['batsman_runs'].count()
x[x>200]
```

```python
x[x>200].index # it will convert all the into the dictionary
list =x[x>200].index.tolist() # and it will convert the whole in list
list
```

```python
y = delivery[delivery['batsman'].isin(list)] # this function will check the batsman coloumn wether the batsman is in the list or not
```

```python
runs = y.groupby('batsman')['batsman_runs'].sum()
balls = y.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr.sort_values(ascending=False).head(3)
```

MERGE : as i am now calculate best betsman in each season
now in matches csv i have season and in deliveries csv i have runs batsman data
in such condition when you have to extract a output by multipe dataset you use merge in pandas

```python
matches = pd.read_csv('matches.csv')
```

```python
new =delivery.merge(matches,left_on='match_id',right_on='id')
new
```

```python
print(matches.shape)
print(delivery.shape)
print(new.shape)
```

```python
final = new.groupby(['season','batsman'])['batsman_runs'].sum().reset_index() # reset index is used for converting a series into dataframe
final
```

```python
final.sort_values('batsman_runs',ascending=False).drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```

Pivot table: it will gives a betteer view according to us to draw conlusions

```python
#we will se the no. 6 in each over by each team
test = delivery[delivery['batsman_runs'] == 6]
pt =test.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
pt
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
