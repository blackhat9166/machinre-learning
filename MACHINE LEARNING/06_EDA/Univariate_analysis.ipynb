```python
import pandas as pd
df = pd.read_csv('tr.csv')
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Catagorical data

- countplot

```python
sns.countplot(df['Survived'])
```

Piechart

```python
#  if you want the same information as above but in percentage mode 
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

2. Numerical data

a. Histogram

```python
plt.hist(df['Age'],bins= 5)
```

b. distplot

```python
sns.distplot(df['Age'])
# curve is KDE kernal density estimation
```

c. Boxplot

```python
sns.boxplot(df['Age'])
```
